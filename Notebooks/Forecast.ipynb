```python
import numpy as np
import pandas as pd
from sklearn import preprocessing, svm
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib import style
import datetime
import math
from sklearn.model_selection import train_test_split

```

```python
forecast_col = 'BCG'
final_df.fillna(value=-99999, inplace=True)
forecast_out = int(math.ceil(0.1 * len(final_df)))
final_df['label'] = final_df[forecast_col].shift(-forecast_out)
print(forecast_out)
final_df['label']
len(final_df)
```

```python
X = np.array(final_df.drop(['label'], 1))
X = preprocessing.scale(X)
# x_lately contains most recent values of x whihc we are going to predict against
X_lately = X[-forecast_out:]
#x is all the rows above x_lately
X = X[:-forecast_out]

#final_df1.dropna(inplace=True)
# define label
y = np.array(final_df['label'][:-2])

print(X,y)
```

```python
# Multiprocessing- program runs various processes and interact, reduces cost, fast processing
# first we take all the data, process it adn split it up
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #trainin gdata is 80% of the data, test is 20% of data.

#LinearReg- model, number of jobs, it looks for the processors that computer have got and work on all of them.
clf = LinearRegression(n_jobs=-1)
#fit the trainin g dtaa in classification model and gives a score
#preprocessing allows for less variance, but takes a lot of time, scale the new and old data, 

clf.fit(X_train, y_train)
#accuracy how many test came out right..
# confidence how confident is the prediction based on % value
accuracy = clf.score(X_test, y_test)
print(accuracy) # VERY SMALL DATASET

confidence = clf.score(X_test, y_test)
print(confidence)
```

```python
# forecast_set is an array of forecasts, showing that not only you
#just seek out a single prediction, but you can seek out many at once.

forecast_set = clf.predict(X_lately)

print(forecast_set, confidence, forecast_out)
```
