```python
#import dependencies
import pandas as pd
import os
import matplotlib
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile

 
```

```python
GNP_data = pd.read_excel('GNPPerCapita.xls', sheet_name='Data')

GNP_data
```

```python
bcg_data = pd.read_excel('immunization_data.xlsx', sheetname='BCG')
bcg_data.head()
```

```python
years_bcg  = list(bcg_data.columns.values[4:])
```

```python
years_gnp  = list(GNP_data.columns.values[4:])
years_gnp
```

```python
bcg_country_code = bcg_data['iso3']
#bcg_country_code
```

```python
data = []
for year in years_bcg:
    for country in bcg_country_code:
        
        gnp_tmp_df = GNP_data[(GNP_data['Country Code'] == country)]
        tmp_gnp = gnp_tmp_df[str(year)]
        bcg_tmp_df = bcg_data[(bcg_data['iso3'] == country)]
        tmp_bcg = bcg_tmp_df[str(year)]
        try:
            data.append({'GNP' : list(tmp_gnp.values)[0], 'BCG': list(tmp_bcg.values)[0]})
        except IndexError:
            pass
            #data.append({'GNP' :0, 'BCG': list(tmp_bcg.values)[0]})
        
#gnp_tmp_df['2017']
```
