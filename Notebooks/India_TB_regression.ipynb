```python
bcg_data = pd.read_excel('immunization_data.xlsx', sheet_name='BCG')
bcg_data.head()
```

```python
#DTP1_data.transpose()
#x_axis_bcg = list(bcg_data)[4:]
bcg_filter= bcg_data[['country','2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009',
       '2008', '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000']]
bcg_filter.head()
```

```python
bcg_arg = bcg_filter.loc[bcg_filter['country']=='India']
bcg_arg.set_index('country')

years_bcg = list(bcg_arg.columns.values[1:])
country= 'India'
years_bcg
```

```python
data = []
for year in years_bcg:
    
        
        
        bcg_tmp_df = bcg_filter[(bcg_filter['country'] == country)]
        tmp_bcg = bcg_tmp_df[str(year)]
        try:
            data.append({'year' : year, 'BCG': list(tmp_bcg.values)[0]})
        except IndexError:
            pass
```

```python
data
bcg_ind = pd.DataFrame(data)
bcg_ind.sort_values(by='year')
```

```python
final_df1 = pd.merge(merged_df, gnp_df, on='year', how='left')
final_df1.to_csv("India_bcg_tb.csv")
#final_df1.set_index('year')
```

```python
from scipy import stats
```

```python
year = final_df1.iloc[:, 0]
bcg_imm = final_df1.iloc[:, 5]
bcg_imm
```

```python
imm_slope, imm_int, imm_r, imm_p, imm_std_err = stats.linregress(
    year, bcg_imm)
imm_fit = imm_slope * year + imm_int
```

```python
gnp_list = final_df1.iloc[:, 6]
gnp_list
```

```python
gnp_slope, gnp_int, gnp_r, gnp_p, gnp_std_err = stats.linregress(
    year, gnp_list)
gnp_fit = gnp_slope * year + gnp_int
```

```python
mort_tb = final_df1.iloc[:, 3]
```

```python
mort_slope, mort_int, mort_r, mort_p, mort_std_err = stats.linregress(
    year, mort_tb)
mort_fit = mort_slope * year + mort_int
```

```python
# Plot
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
fig.suptitle("BCG Immunization, Mortality, and GNP of India", fontsize=10, fontweight="bold")

ax1.set_xlim(min(year), max(year))
ax1.plot(year, bcg_imm, linewidth=1, marker="o", color = "b")
ax1.plot(year, imm_fit, "b--", linewidth=1)
ax1.set_ylabel("% BCG")

ax2.plot(year, gnp_list, linewidth=1, marker="o", color="r")
ax2.plot(year, gnp_fit, "r--", linewidth=1)
ax2.set_ylabel("GNP")

ax3.plot(year, mort_tb, linewidth=1, marker="o", color="g")
ax3.plot(year, mort_fit, "g--", linewidth=1)
ax3.set_ylabel("TB Mortality")
ax3.set_xlabel("Year")
plt.savefig('Ind_Immunization.png')
```
