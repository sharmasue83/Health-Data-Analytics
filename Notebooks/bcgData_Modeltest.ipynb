```python
#import dependencies
import pandas as pd
import os
import matplotlib
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
country_list = bcg_data['country']
```

```python
new_df = {'Years':list(bcg_data.columns.values[4:]),
         '%Immunization': bcg_data.loc[bcg_data['country']== 'Argentina']}
```

```python
Years=list(bcg_data.columns.values[4:])
Years
#Immunization= bcg_data.loc[bcg_data['country']== 'Argentina']
#Immunization
```

```python
data = []
for year in Years:
    for country in country_list:
    
        bcg_tmp_df = bcg_data[(bcg_data['country'] == country)]
        tmp_bcg = bcg_tmp_df[str(year)]
        data.append({'Year' : int(year), 'BCG': list(tmp_bcg.values)[0]})
        
            #data.append({'GNP' :0, 'BCG': list(tmp_bcg.values)[0]})
data   
```

```python

bcg_argentina =pd.DataFrame(data)
bcg_argentina.dropna(inplace=True)
```

```python

import numpy as np
from sklearn import preprocessing, svm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


```

```python
# considering year as label as we need to predict the year
X = np.array(bcg_argentina.drop(['Year'], 1))
y = np.array(bcg_argentina['Year'])



```

```python
# Preprocessing scales the feature set
# 

X = preprocessing.scale(X)
y = np.array(bcg_argentina['Year'])
```

```python
#Takes 2 arrays, gives 4 variabes... test size is 20 % of train... 
# train data is selected based built-in algorithm
#allows for re-training... if data become bigger, change parameters and re-train...

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Support VEctor regression

clf = svm.SVR()
clf.fit(X_train, y_train)

confidence = clf.score(X_test, y_test)
print(confidence)
```
